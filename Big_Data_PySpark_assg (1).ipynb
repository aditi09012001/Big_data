– TASK-1: Generate 100 random numbers in range 0 to 10 using numpy

randint function with the seed set to 10. Create a RDD using the paral-
lelize function using data generated in previous step. Calculate the fre-
quency of each number (0 - 10) using appropriate function of RDD.

```python
# Initialize Spark Context
sc = SparkContext(appName="NumberFrequency")

# Set seed and generate 100 random numbers between 0 and 10
np.random.seed(10)
random_numbers = np.random.randint(0, 11, 100)

# Create RDD from the generated numbers
numbers_rdd = sc.parallelize(random_numbers)

# Calculate the frequency of each number
frequency = numbers_rdd.countByValue()

# Sort the frequency dictionary by key for better readability
sorted_frequency = dict(sorted(frequency.items()))

# Displaying the frequency of each number with a new format
print("Number\t|\tFrequency")
print("-" * 25)
for number, freq in sorted_frequency.items():
    print(f"{number}\t|\t{freq}")

# Stop the Spark Context
sc.stop()
```

```python
# Source URL for acquiring the text data
data_source = "http://mattmahoney.net/dc/text8.zip"

# Setting the target directory for storing the compressed file
storage_directory = "./dataset/"

# Ensuring the existence of the specified directory
Path(storage_directory).mkdir(parents=True, exist_ok=True)
compressed_file_path = Path(storage_directory, "text8_compressed.zip")

extracted_data_dir = Path(storage_directory, "text8_data/")

# Ensuring the designated extraction directory exists
extracted_data_dir.mkdir(parents=True, exist_ok=True)

# Fetching and downloading the text data from the provided source URL
data_response = requests.get(data_source)
compressed_file_path.write_bytes(data_response.content)

# Unzipping and extracting the contents of the downloaded file
with zipfile.ZipFile(compressed_file_path, 'r') as data_zip_file:
    data_zip_file.extractall(extracted_data_dir)

print("Text data downloaded and extraction completed successfully.")
```

```python
# Specifying the path to the text file within the extracted folder
text_file_path = extracted_data_dir / "text8"

# Reading the contents of the text file
with open(text_file_path, 'r', encoding='utf-8') as file:
    data_text = file.read()

# Printing the 500 characters of the text data
print(data_text[200:700])

# Stopping the Spark Context
sc.stop()
```

```python
# Initiating the Spark Context for Covert Operations
covert_context = SparkContext("local", "CovertOperationRDD")

# Creating an RDD with encoded confidential information split into lines
encoded_info_rdd = covert_context.parallelize(data_text[200:700].splitlines())

# Splitting each line into fragments and filtering those with the secret letter 'a'
concealed_fragments_rdd = encoded_info_rdd.flatMap(lambda line: line.split(" ")).filter(lambda fragment: 'a' in fragment)

# Mapping each concealed fragment to a cryptic tuple with the fragment as the key and an undisclosed value of 1
encrypted_fragments_rdd = concealed_fragments_rdd.map(lambda fragment: (fragment, 1))

# Decrypting by key to reveal the frequency of each covert fragment
decrypted_frequencies_rdd = encrypted_fragments_rdd.reduceByKey(lambda x, y: x + y)

# Gathering the results discreetly
covert_results = decrypted_frequencies_rdd.collect()

# Displaying the covert frequencies covertly
print("Covert Frequencies:")
for covert_code, occurrences in covert_results:
    print(f"Covert Code: {covert_code}, Occurrences: {occurrences}")

# Concluding the Covert Operations Spark Context to maintain secrecy
covert_context.stop()
```

• DataFrame Task: Create a Spark dataframe using the iris json data men-
tioned above. Calculate Pearson Correlation between the columns petal-
Length and petalWidth using the appropriate dataframe API. Show the

columns sepalLength, sepalWidth and species for the rows of data that has
petalLength greater than or equal to 1.4 using the appropriate dataframe

```python
ref_zip = zipfile.ZipFile('/content/archive (1).zip', 'r')
ref_zip.extractall('/content')
ref_zip.close()
```

```python
iris_df=pd.read_csv('/content/iris.json')
iris_df.head(10)
```

```python
# Initializing Spark environment for the Floral DataFrame Task
floral_spark = SparkSession.builder.appName("FloralDataFrameTask").getOrCreate()

# Defining the schema for the iris dataset
iris_schema = StructType([
    StructField("sepalLength", DoubleType(), True),
    StructField("sepalWidth", DoubleType(), True),
    StructField("petalLength", DoubleType(), True),
    StructField("petalWidth", DoubleType(), True),
    StructField("species", StringType(), True),
])

# Loading the iris dataset from the JSON file
iris_df = floral_spark.read.json("/content/iris.json", schema=iris_schema)

# Displaying the first few rows of the Iris DataFrame
iris_df.show(5)

# Calculating the Pearson Correlation between petalLength and petalWidth
petal_correlation = iris_df.stat.corr("petalLength", "petalWidth")
print(f"Pearson Correlation between petalLength and petalWidth: {petal_correlation:.4f}")

# Displaying selected columns for rows with petalLength >= 1.4
selected_floral_df = iris_df.filter(col("petalLength") >= 1.4).select("sepalLength", "sepalWidth", "species")
selected_floral_df.show()

# Stopping the Spark session
floral_spark.stop()
```
